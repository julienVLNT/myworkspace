## Naviguer dans l'arborescence depuis le script.

En langage Python, la librairie standard inclut le package `os`, permettant d'interagir avec l'arborescence.

```python
import os
```

On accède au chemin absolu du script en train d'être exécuté à l'aide de `os.getcwd()`. 

```python
os.getcwd()
```

On peut ensuite parcourir récursivement un dossier à l'aide de la structure suivante.

```python
for root, dirs, files in os.walk (".", topdown=True, onerror=None, followlinks=False):
    for name in files:
        print (f"Type: file - {os.path.join (root, name)}")
    for name in dirs:
        print (f"Type: dir  - {os.path.join (root, name)}")
        
    print (80 * "=")
```
