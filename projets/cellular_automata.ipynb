# Notes sur les automates cellulaires et leur classification par Stephen Wolfram

## Implémentation

### Environnement de développement

```python
import matplotlib.pyplot as plt
import numpy             as np
```

### Classe `Automaton`

```python
class Automaton():
    """Cellular automaton.
    """


    def __init__(self, length: int, rule: int) -> None:
        """Constructor.
        
        Parameters
        ----------
        length    : int,
                    The size of the vector that represents the state.
        rule      : int [0, 255]
                    The decimal code of the rule to compute the next step.
        
        Return
        ------
        None
        """
        # Initialisation of the states
        self.states = np.zeros((1, length), dtype=np.int8)
        
        # Build of the rule as a dictionary
        states = ['111', '110', '101', '100', '011', '010', '001', '000']
        rule = [c for c in bin(rule)[2:].zfill(8)]
        self.law = dict(zip(states, rule))
        
        return None
    
    
    def initialise(self, state: np.ndarray) -> None:
        """Set the initial state of the automaton.
        
        Parameters
        ----------
        state     : ndarray
                    The state to set in self.states[0].
        Return
        ------
        None
        """
        self.states[0] = state
        
        return None


    def plot(self) -> None:
        """Plots the image of the temporal evolution of the automaton.
        """
        Ny, Nx = self.states.shape
        
        fig = plt.figure(figsize=(Ny, Nx))
        
        axe = fig.add_subplot(1, 1, 1)
        axe.imshow(self.states, origin='lower')
        axe.set_title("Evolution")
        axe.set_xlabel("X [cell]")
        axe.set_ylabel("T [index]")

        plt.show()
        
        return None
    
    
    def show(self) -> None:
        """Prints the current state of the automaton.
        """
        print(self.states[-1])
        
        return None
    
    
    def step(self) -> None:
        """Computes and stores in self.states the next state.
        """
        
        new = np.zeros_like(self.states[-1])
        
        for k in range(self.states.shape[1]):
            if k == self.states.shape[1]-1:
                current = ''.join([str(i) for i in self.states[-1, -2:]] + [str(self.states[-1, 0])])
                new[k]  = self.law[current]
                
            elif k == 0:
                current = ''.join([str(self.states[-1, -1])] + [str(i) for i in self.states[-1, :2]])
                new[k]  = self.law[current]
            
            else:
                current = ''.join([str(i) for i in self.states[-1, k-1:k+2]])
                new[k]  = self.law[current]
        
        self.states = np.vstack((self.states, new))
        
        return None

    
    def evolve(self, n: int) -> None:
        """Computes n steps of the evolution.

        Parameters
        ----------
        n         : int
                    Number of timesteps to compute.
        
        Return
        ------
        None
        """
        for _ in range(n):
            self.step()
        
        return None
```

## Application : dessiner le triangle de Sapierski

```python
length    = 200
times     = 200
sapierski = Automaton(length, 60)
sapierski.initialise(np.random.randint(0, 2, length))
sapierski.evolve(times)
sapierski.plot()
```
